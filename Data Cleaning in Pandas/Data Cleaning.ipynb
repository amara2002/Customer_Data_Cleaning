# Dropping Duplicates

```python
df = df.drop_duplicates()
df
```

# Drop Not Useful Column

```python
df = df.drop(columns = "Not_Useful_Column")
```

```python
df
```

# Cleaning Last Name

```python
df["Last_Name"] = df["Last_Name"].str.strip("123../")
```

```python
df
```

# Cleaning Phone Number

```python
def clean_phone(phone):
    # Convert to string
    phone_str = str(phone).strip()
    
    # Treat invalid or missing entries as NaN
    if phone_str.lower() in ['', 'nan', 'n/a', 'na']:
        return np.nan
    
    # Keep only digits
    digits = ''.join(filter(str.isdigit, phone_str))
    
    # If not enough digits, return as is or NaN
    if len(digits) < 10:
        return np.nan
    
    # Format: 123-545-5421 (first 3, next 3, last 4)
    formatted = f"{digits[:3]}-{digits[3:6]}-{digits[6:10]}"
    return formatted

# Apply the function
df['Phone_Number'] = df['Phone_Number'].apply(clean_phone)

df
```

# Expanding Street

```python
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', n=2, expand=True)
```

```python
df
```

# Replacing No with 'N' and Yes with 'Y'

```python
df["Paying Customer"] = df["Paying Customer"].replace({'Yes':'Y', 'No':'N'})
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace({'Yes':'Y', 'No':'N'})
```

```python
df
```

# Replace Not a value with ""

```python
df = df.fillna("")
```

```python
df
```

# Drop Y in Do Not Contact

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == '':
        df.drop(x, inplace=True)
```

```python
df
```

# Removing "" Rows in Phone Number

```python
for y in df.index:
    if df.loc[y, "Phone_Number"] == "":
      df.drop(y, inplace=True)
```

```python
df
```

# Resetting the Table

```python
df = df.reset_index(drop=True)
```

```python
df
```
